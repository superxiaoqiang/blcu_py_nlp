```python
import requests
from pyquery import PyQuery as pq
url = 'http://www.blcu.edu.cn/'
```

```python
r=requests.get(url)
html=r.text.encode(r.encoding).decode(r.apparent_encoding,errors='ignore')
```

```python
doc=pq(html)
lis = doc('.zcontent li')
```

```python
lis
```

```python
for li in lis:
    print(pq(li).find('a').text())
```
